```python
import pandas as pd
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', None)  # or 199
```

```python
import glob
dir_path = r'daily_trading_report\*'
daily_trading_csv_list = [] 
for file in glob.glob(dir_path, recursive=True):
    daily_trading_csv_list.append(file)
    print(file)
```

```python
# concat all daily trading record csv
df = pd.concat(map(pd.read_csv, daily_trading_csv_list))
```

```python
df['Symbol'] = df.apply(lambda x: x['DESCRIPTION'].split()[-2], axis=1)
df['Buy / Sell'] = df.apply(lambda x: x['DESCRIPTION'].split()[0], axis=1)
df['Shares'] = df.apply(lambda x: x['DESCRIPTION'].split()[1], axis=1)
```

```python
# reorder header
cols = df.columns.tolist() 
cols = ['DATE','TIME','Symbol','Buy / Sell', 'Shares','TYPE', 'DESCRIPTION', 'Misc Fees', 'AMOUNT','BALANCE',]
df = df[cols]
```

```python
# df
# df['Holding Time'] = df.apply(lambda x: x['DESCRIPTION'].split()[1], axis=1)
df.sort_values(['Symbol', 'TIME'], ascending=[True, True], inplace=True)
```

```python
df['Profit'] = 0 
df['Holding Time'] = None 
df['TIME'] = pd.to_datetime(df['TIME'])
```

```python
df
```

```python
df['AMOUNT'] = df.apply(lambda x: str(x['AMOUNT']).replace(',', ''), axis=1)
df['AMOUNT'] = df['AMOUNT'].astype(float)
```

```python
df = df.reset_index(drop=True)
for i in range(1, len(df)):
    if df.loc[i]["Symbol"] == df.loc[i-1, "Symbol"] and df.loc[i, "Buy / Sell"] == "SOLD" and df.loc[i-1, "Buy / Sell"] == "BOT":
        df.loc[i, 'Profit'] = df.loc[i, 'AMOUNT'] + df.loc[i-1, 'AMOUNT']
        df.loc[i, 'Holding Time'] = pd.Timedelta(df.loc[i, 'TIME'] - df.loc[i-1, 'TIME'])
```
